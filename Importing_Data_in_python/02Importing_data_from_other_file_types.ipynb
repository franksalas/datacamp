````python
# Import pandas
import pandas as pd

# Assign spreadsheet filename: file
file = 'battledeath.xlsx'

# Load spreadsheet: xl
xl = pd.ExcelFile(file)

# Print sheet names
print(xl.sheet_names)
````

````python
# Load a sheet into a DataFrame by name: df1
df1 = xl.parse('2004')

# Print the head of the DataFrame df1
print(df1.head())

# Load a sheet into a DataFrame by index: df2
df2 = xl.parse(0)

# Print the head of the DataFrame df2
print(df2.head())
````

````python
# Parse the first sheet and rename the columns: df1
df1 = xl.parse(0, skiprows=[0], names=['Country','AAM due to War (2002)'])

# Print the head of the DataFrame df1
print(df1.head())

# Parse the first column of the second sheet and rename the column: df2
df2 = xl.parse(1, parse_cols=[0], skiprows=[0], names=['Country'])

# Print the head of the DataFrame df2
print(df2.head())
````

# Importing HDF5 files

## HDF5 files
- Hierarchical Data Format version 5
- Standard for storing large quantities of numerical data
- Datasets can be hundreds of gigabytes or terabytes
- HDF5 can scale to exabytes

## importing HDF5 files

````python
import h5py
filename = 'L-L1_LOSC_4_V1-1126259446-32.hdf5'

data = h5py.File(filename, 'r')  # 'r' is to read

type(data)
````

## The structure of HDF5 files

````python
for key in data.keys():
    print(key)
````

````python
type(data['meta'])
````

````python
for key in data['meta'].keys():
    print(key)
````

````python
data['meta']['Description'].value, data['meta']['Detector'].value
````

# Let's practice!

````python
# Import packages
import numpy as np
import h5py

# Assign filename: file
file = 'L-L1_LOSC_4_V1-1126259446-32.hdf5'

# Load file: data
data = h5py.File(file, 'r')

# Print the datatype of the loaded file
print(type(data))

# Print the keys of the file
for key in data.keys():
    print(key)
````

````python
# Get the HDF5 group: group

group = data['strain']

# Check out keys of group
for key in group.keys():
    print(key)

# Set variable equal to time series data: strain
strain = data['strain']['Strain'].value

# Set number of time points to sample: num_samples
num_samples = 10000

# Set time vector
time = np.arange(0, 1, 1/num_samples)

# Plot data
plt.plot(time, strain[:num_samples])
plt.xlabel('GPS Time (s)')
plt.ylabel('strain')
plt.show()
````

- Industry standard in engineering and science
- Data saved as .mat files

## SciPy to the rescue!
- `scipy.io.loadmat()` - read .mat files
- `scipy.io.savemat()` - write .mat files

## Importing a .mat file

- keys = MATLAB variable names
- values = objects assigned to variables
